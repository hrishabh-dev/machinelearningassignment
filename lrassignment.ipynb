### 1  What is Logistic Regression, and how does it differ from Linear Regression.

```python
'''

Logistic Regression is a statistical method used for binary classification problem

Logistic Regression: Assumes a non-linear relationship  between the dependent variable
and independent variable(s).

the goal is to predict the probability of an outcome that can take one of two possible values
0 OR 1 

It models the relationship between a dependent binary variable and one or 
more independent variables 

Linear Regression: Produces continuous outcomes. It predicts a real-valued number

Linear Regression: The predicted value can be any real number

Linear Regression: Used for regression tasks, like predicting sales, stock prices

Linear Regression: Assumes a linear relationship between the independent and dependent variable

'''
```

### 3 Why do we use the Sigmoid function in Logistic Regression.

```python
'''

The sigmoid function naturally squashes any real-valued number into the range of 0 and 1

The output of the sigmoid function can be interpreted as a probability

The sigmoid function has an S-shaped (or "S-curve") form, which means that it approaches 0 and 1

The sigmoid function is smooth and differentiable, which is essential for optimizing
logistic regression models.

the sigmoid function is crucial in logistic regression for transforming linear combinations
of input features into interpretable probabilities, providing a probabilistic
framework for binary classification while maintaining necessary mathematical
properties for effective optimization

'''
```

### 5 What is Regularization in Logistic Regression? Why is it needed.

```python
'''
Regularization in logistic regression  is a technique used to prevent overfitting
by introducing additional information or constraints into the model. 
It involves adding a penalty term to the cost function

Needed :-

Overfitting occurs when a model learns the noise in the training
data rather than the underlying pattern.

Regularization can lead to more stable and robust estimates of the model parameters.

Regularization can aid in feature selection

echniques such as L1 (Lasso) regularization can drive some coefficients to exactly zero,


'''
```

### 6 Explain the difference between Lasso, Ridge, and Elastic Net regressionC

```python
'''
1. Lasso Regression (L1 Regularization)

Penalty Term: Lasso adds an L1 penalty to the cost function, which
is the sum of the absolute values of the coefficients.

 Lasso can shrink some coefficients down to exactly zero, effectively performing feature selection. 

  Lasso is particularly effective when you suspect that only a few features are actually important


2. Ridge Regression (L2 Regularization)

Penalty Term: Ridge adds an L2 penalty to the cost function, which is 
the sum of the squares of the coefficients.

it does not set coefficients to exactly zero.

 Ridge is particularly useful when there are many small/medium effect features

3. Elastic Net

Penalty Term: Elastic Net combines both L1 and L2 penalties in the cost function

 Elastic Net seeks to balance the benefits of both Lasso  and Ridge

 Elastic Net is often used when you have a larger dataset with lots of correlated features 

'''
```

### 7 When should we use Elastic Net instead of Lasso or Ridge.

```python
'''

Elastic Net is a hybrid regularization technique that combines the properties
of both Lasso (L1 regularization) and Ridge (L2 regularization)


It is particularly useful in specific situations where the limitations of
Lasso and Ridge can be addressed


elastic Net can retain these correlated features by using both the L1 and L2 penalties


Elastic Net can offer more stable solutions by using both penalty types


 Elastic Net allows you to leverage the strengths of both.

 It can help retain both features and their interactions
 through the grouping effect enabled by Ridge's penalty.
'''
```

### 8 What is the impact of the regularization parameter (λ) in Logistic Regression.

```python
'''

 the regularization parameter (λ or "lambda") plays a crucial role in controlling the complexity
 of the model by applying regularization techniques

 the impact of the regularization parameter on logistic regression:

Large λ: A higher value of λ increases the penalty for large coefficients,

 less important features can be completely excluded from the model, resulting 
 in a sparser and more interpretable model.

 Higher bias, lower variance: Increasing λ typically increases bias 
 but decreases variance 

  the selection of λ influences the model's predictive performance.

  A well-chosen λ can enhance generalization capabilities
 

'''
```

### 9 What are the key assumptions of Logistic Regression.

### 15  How is Logistic Regression extended for multiclass classification.

```python
'''

it can be extended to handle multiclass classification problems, where
there are three or more classes to predict

. One-vs-Rest (OvR) or One-vs-All (OvA)

 In the One-vs-Rest approach, a separate binary logistic regression
 model is trained for each class.

 the model predicts the probability of the instance belonging to
 that class versus all other classes combined.

 2. Softmax Regression (Multinomial Logistic Regression)
 
Softmax regression is an extension of logistic regression that allows
the model to directly predict multiple classes with a single model. 

If interpretability is crucial and the dataset is smaller, OvR might be preferred.

For larger datasets  Softmax regression is typically more suitable.


'''
```

### 16 What are the advantages and disadvantages of Logistic Regression.

```python
'''
Advantages of Logistic Regression

Simplicity and Interpretability:
Logistic regression is straightforward to understand and easy to implement.

Efficiency:

It is computationally inexpensive compared to more complex algorithms, making it suitable

Probabilistic Outputs:

Logistic regression provides class probabilities, which can be useful for applications

egularization:

Logistic regression can be easily regularized (using L1 or L2 regularization), which 
helps prevent overfitting

Disadvantages of Logistic Regression

Linearity Assumption:

Logistic regression assumes a linear relationship between the 
independent variables and the log-odds of the outcome. 

Sensitive to Outliers:

Logistic regression can be sensitive to outliers and influential data points, which can 
distort the estimated coefficients significantly.

Limited Flexibility:

The linearity assumption limits the model's flexibility, particularly in capturing complex 
relationships in larger and richer datasets.


'''
```

### 18 What is the difference between Softmax Regression and Logistic Regression.

```python
'''

Logistic Regression: 

Typically used for binary classification tasks. It predicts the probability 
of a data point belonging to one of two classes

The model outputs a single probability score ranging from 0 to 1 for the positive class

Suitable for binary outcomes and simpler implementations. It is computationally 
efficient and can be directly applied

Softmax Regression:

Used for multiclass classification tasks, where there are more than two classes

The model outputs a probability distribution across all classes, where the 
sum of the probabilities for all classes equals 1

Used in scenarios requiring classification into multiple classes.

'''
```

### 19  How do we choose between One-vs-Rest (OvR) and Softmax for multiclass classification

```python
'''

If your classification problem is relatively simple, and you already have a binary
classifier that works well, OvR can be used 

 In cases where the number of classes is large, and some classes have 
very few samples, OvR can allow you to build classifiers

OvR creates separate models for each class, it can provide more interpretable results

a unified model with a single training process, Softmax is preferable.

 Softmax regression typically requires less computational overhead


  Softmax can better capture the relationships between different classes

'''
```

### 20 How do we interpret coefficients in Logistic Regression?

```python
'''
Each coefficient βi represents the change in the log odds of the
outcome variable Y in relation to a one-unit increase in the predictor 
Xi while holding other predictors constant.

coefficients do not directly give you probabilities but affect the predicted 
probabilities through the logistic function.

If a coefficient β2=-0.3 for another predictor X2

 it implies that for each unit increase in income, the odds 
 of making a purchase decrease by a factor of e**−0.3≈0.74 (or about 26% decrease), 
 holding all other factors constant.

'''
```

### 7 Write a Python program to evaluate Logistic Regression using Stratified K-Fold Cross-Validation. Print the average accuracy

```python
from sklearn.model_selection import StratifiedKFold
```

```python
iris = load_iris()
X = iris.data   
y = iris.target 
sk = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
sk
```

```python
log_reg = LogisticRegression(solver='liblinear')
accuracy_l = []
for train_index, test_index in sk.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    log_reg.fit(X_train, y_train)
    y_pred = log_reg.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_l.append(accuracy)
```

```python
average_accuracy = np.mean(accuracy_l)
print(f"Average Accuracy from Stratified K-Fold Cross-Validation: {average_accuracy:.4f}")
```

### 25  Write a Python program to train Logistic Regression, save the trained model using joblib, and load it again to make predictions.

```python
import joblib
```

```python
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
model = LogisticRegression(solver='lbfgs', multi_class='auto', max_iter=200)
model.fit(X_train, y_train)
```

```python
joblib_file = "logistic_regression_model.joblib"
joblib.dump(model, joblib_file)
print(f'Model saved to {joblib_file}.')
```

```python
load_model = joblib.load(joblib_file)
print('Model loaded successfully.')
```

```python
y_pred = load_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of the loaded model: {accuracy:.4f}')
```
